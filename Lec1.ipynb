```python
#Importing packages and Mounting Google Drive to Colab
import pandas as pd 
from google.colab import drive
import matplotlib.pyplot as plt
drive.mount('/content/gdrive',force_remount =True) 
```

```python
#Reading in the data
df = pd.read_csv("/content/gdrive/MyDrive/adult.csv")
df
```

```python
#Visualizing the data
plt.figure(figsize=(8,6))
plt.hist(df["hours_per_week"])
plt.title("Census Data: hours worked per week")
plt.xlabel("No. of hours")
plt.ylabel("No. of people")
```

The above histogram shows the number of hours worked per week on the x-axis and the corresponding number of people on the y-axis from US Census Data. We see from the plot that the mode lies around 30 hours per week and the distribution has a slight positive skew.
